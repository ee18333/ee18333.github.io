https://api.covid19api.com/total/country/us/status/deaths

```python
url_base = "https://api.covid19api.com/total/country/{}/status/deaths"
file_base = "data_covid19api-{}.json"
countryname = ['uk', 'us', 'france', 'germany', 'spain', 'norway', 'sweden']
for i in countryname: 
   URL = url_base.format(i)
   print(URL)

import requests

import os

import json

from google.colab import files

data = requests.get(URL).json()
print(data)

fileName = file_base.format(i)
print(fileName)

with open(fileName, 'w', encoding='utf-8') as f:
  json.dump(data, f, ensure_ascii=False, indent=4)

files.download(fileName)
```
